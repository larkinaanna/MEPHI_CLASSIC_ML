```python
def filter_low_variance_features(dataframe, min_variance=0.1, verbose=True):
    """
    Исключает признаки с дисперсией ниже заданного порога.

    Аргументы:
    - dataframe: pd.DataFrame - исходные данные
    - min_variance: float - минимально допустимая дисперсия (по умолчанию 0.1)
    - verbose: bool - выводить информацию об удалённых признаках

    Возвращает:
    - pd.DataFrame - данные с отфильтрованными признаками
    """
    # Вычисляем дисперсию для каждого признака
    variances = dataframe.var()

    # Определяем признаки, которые нужно сохранить
    keep_mask = variances > min_variance
    selected_features = dataframe.columns[keep_mask]

    # Формируем список удаляемых признаков
    dropped_features = [col for col in dataframe.columns if col not in selected_features]

    if verbose and dropped_features:
        print(f"Удалены признаки с низкой дисперсией (≤{min_variance}): {', '.join(dropped_features)}")

    # Возвращаем DataFrame только с отобранными признаками
    return dataframe[selected_features].copy()

df_final = filter_low_variance_features(df)
```

```python
def eliminate_correlated_features(data, correlation_threshold=0.9):
    """
    Удаляет сильно коррелирующие признаки, сохраняя только один из каждой группы.

    Параметры:
    - data: pd.DataFrame - входной датафрейм
    - correlation_threshold: float - порог для определения высокой корреляции (0.9 по умолчанию)

    Возвращает:
    - pd.DataFrame - датафрейм без скоррелированных признаков
    """
    # Вычисляем матрицу абсолютных корреляций
    correlation_matrix = data.corr().abs()

    # Создаем множество для хранения удаляемых признаков
    features_to_remove = set()

    # Проходим по верхнему треугольнику матрицы корреляций
    for i in range(len(correlation_matrix.columns)):
        for j in range(i):
            if correlation_matrix.iloc[i, j] > correlation_threshold:
                colname = correlation_matrix.columns[i]
                features_to_remove.add(colname)

    # Преобразуем в список для стабильного порядка
    removed_features = sorted(features_to_remove)

    if removed_features:
        print("Удалены следующие коррелирующие признаки:")
        print(removed_features)

    # Возвращаем данные без удаленных признаков
    return data.drop(columns=removed_features)

df_final = eliminate_correlated_features(df_final)
```

```python
corr_matrix = df_final.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Корреляционная матрица")
plt.show()
```
